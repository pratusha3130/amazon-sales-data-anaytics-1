```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/FDI data.csv")
```

```python
df.head()
```

```python
df.set_index("Sector", inplace=True)
```

```python
df = df.transpose()
```

```python
print ('Descriptive Statistics of FDI data')
df.describe()
```

```python
df['Total FDI'] = df.sum(axis=1)
yearly_fdi = df['Total FDI']
plt.figure(figsize=(10, 6))
yearly_fdi.plot(kind='bar')
plt.title('Total FDI for Each Year')
plt.xlabel('Year')
plt.ylabel('Total FDI')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(data = yearly_fdi)
plt.title('Total FDI for Each Year')
plt.xlabel('Year')
plt.ylabel('Total FDI')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
growth_rates = yearly_fdi.pct_change() * 100
plt.figure(figsize=(10, 8))
sns.lineplot(data = growth_rates)
plt.title('Growth Rates of FDI')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
growth_rate = df.pct_change(axis = 1) * 100
plt.figure(figsize=(10, 8))
sns.lineplot(data = growth_rate)
plt.title('Growth Rates of FDI')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=60)
plt.show()
```

```python
# Top 5 Best Performing years
sorted = df.sort_values(by='Total FDI', ascending=False)
top_five = sorted.head(5)
top_five
```

```python
# Top 5 Categories
newdf= df.transpose()
newdf.drop(index='Total FDI', inplace=True)
newdf['Category Total'] = newdf.sum(axis=1)
sorted = newdf.sort_values(by='Category Total', ascending=False)
top_five = sorted.head(5)
top_five
```

```python
top5_categories = newdf.head(5)
top5_categories = pd.DataFrame(top5_categories)
top5_categories
```

```python
top5_categories= top5_categories.transpose()
```
